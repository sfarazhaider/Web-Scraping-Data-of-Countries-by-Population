## Importing Libraries

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests 
```

## Creating soup

```python
url = 'https://www.worldometers.info/world-population/population-by-country/'
```

```python
page=requests.get(url)
```

```python
soup= BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

## Filtering Headers 

```python
soup.find_all('th')
```

```python
world_titles = soup.find_all('th')
world_titles
```

```python
header= [title.text.strip() for title in world_titles]
header
```

## Creating Data Frame

```python
df = pd.DataFrame(columns=header)
df
```

```python
tabledata = soup.find_all('tr')
tabledata
```

```python
for row in tabledata[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    ##print(individual_row_data)

    length = len(df)
    df.loc[length] = individual_row_data
    
```

```python
df
```
